```python
#Number-Guessing Game

import random

# Function to check if the guess is a valid 5-digit number with no repeated digits
def is_valid_number(number):
    if len(number) != 5 or not number.isdigit():
        return False
    if len(set(number)) != 5:
        return False
    return True

# Function to get the secret number from the administrator
def get_secret_number():
    while True:
        secret = input("Enter the secret 5-digit number (no repeating digits): ")
        if is_valid_number(secret):
            print("\n" *5)  # Clear the screen to hide the secret number
            return secret
        else:
            print("Invalid input. Please enter a valid 5-digit number with no repeating digits.")

# Function to check the guess and give feedback
def check_guess(secret, guess):
    correct_digits = 0
    correct_positions = 0
    for i in range(5):
        if guess[i] in secret:
            correct_digits += 1
        if guess[i] == secret[i]:
            correct_positions += 1
    return correct_digits, correct_positions

# Main function to run the game
def main():
    # Option to get secret number from admin or generate randomly
    choice = input("Would you like to (1) Provide the secret number or (2) Generate a random one? (Enter 1 or 2): ")
    if choice == '1':
        secret = get_secret_number()
    else:
        secret = ''.join(random.sample('0123456789', 5))  # Random 5-digit number with no repeating digits
        print("\n" * 50)  # Hide the secret number for the player

    max_guesses = 10
    guess_count = 0

    while guess_count < max_guesses:
        guess = input(f"Guess #{guess_count + 1} (Enter a 5-digit number): ")

        # Validate the guess
        if not is_valid_number(guess):
            print("Invalid guess. Please enter a valid 5-digit number with no repeating digits.")
            continue

        guess_count += 1
        correct_digits, correct_positions = check_guess(secret, guess)

        print(f"Feedback: {correct_digits} correct digits, {correct_positions} correct positions")

        if correct_positions == 5:
            print(f"Congratulations! You've guessed the secret number {secret} in {guess_count} guesses.")
            break
    else:
        print(f"Sorry, you've used all {max_guesses} attempts. The secret number was {secret}.")

    print("Game over!")

if _name_ == "_main_":
    main()
```
