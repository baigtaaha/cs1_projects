```python
#CarRental
def get_classification_code():
    while True:
        # Ask the user for the classification code
        code = input("Enter the customer's classification code (B, D, W or Q to quit): ").upper()
        if code in ['B', 'D', 'W', 'Q']:  # Check if the input is valid
            return code
        else:
            print("Invalid classification code. Please enter 'B', 'D', 'W', or 'Q'.")
def get_number_input(prompt):
    while True:
        try:
            # Ask for an integer input and ensure it's a valid number
            return int(input(prompt))
        except ValueError:
            print("Invalid input! Please enter a valid number.")
def calculate_miles_driven(start, end):
    # Calculate the miles driven, accounting for rollover of the odometer
    if end >= start:
        return (end - start) / 10.0
    else:
        return ((1000000 - start) + end) / 10.0
def calculate_charge(code, days, miles):
    # Calculate the charge based on the classification code
    if code == 'B':  # Budget
        base_charge = 40 * days
        mileage_charge = 0.25 * miles
        total_charge = base_charge + mileage_charge
    elif code == 'D':  # Daily
        base_charge = 60 * days
        if miles / days > 100:  # Charge extra if average miles/day is above 100
            mileage_charge = 0.25 * (miles - 100 * days)
        else:
            mileage_charge = 0
        total_charge = base_charge + mileage_charge
    elif code == 'W':  # Weekly
        base_charge = 190 * ((days - 1) // 7 + 1)  # Full week charge
        if miles / 10 > 900:
            if miles / 10 > 1500:
                mileage_charge = 200 + 0.25 * (miles - 1500 * 10)
            else:
                mileage_charge = 100
        else:
            mileage_charge = 0
        total_charge = base_charge + mileage_charge
    else:
        total_charge = 0  # Default case (not needed since 'Q' will stop the loop)
    return total_charge
def main():
    while True:
        classification = get_classification_code()
        if classification == 'Q':  # Exit the program if 'Q' is entered
            print("Exiting the program.")
            break
        days = get_number_input("Enter the number of days the vehicle was rented: ")
        start_odometer = get_number_input("Enter the starting odometer reading: ")
        end_odometer = get_number_input("Enter the ending odometer reading: ")
        # Calculate miles driven
        miles = calculate_miles_driven(start_odometer, end_odometer)
        # Calculate total charge
        total_amount = calculate_charge(classification, days, miles)
        # Display the result
        print(f"\nRental Summary:")
        print(f"Classification code: {classification}")
        print(f"Days rented: {days}")
        print(f"Starting odometer reading: {start_odometer}")
        print(f"Ending odometer reading: {end_odometer}")
        print(f"Miles driven: {round(miles, 1)} miles")
        print(f"Total amount billed: ${total_amount:.2f}\n")
if _name_ == "_main_":
    main()
```
