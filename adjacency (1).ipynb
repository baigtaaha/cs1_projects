```python
#Adjacency

class Matrix:
    def _init_(self):
        self._rooms = 0
        self._matrix = []

    def read_file(self, file):
        """Reads the file to create the adjacency matrix."""
        self._rooms = int(file.readline())  # Read number of rooms
        self._matrix = [set() for _ in range(self._rooms)]

        for line in file:
            parts = line.strip().split()
            if len(parts) == 2:
                try:
                    src, dst = map(int, parts)
                    if 0 <= src < self._rooms and 0 <= dst < self._rooms:
                        self._matrix[src].add(dst)
                        self._matrix[dst].add(src)
                    else:
                        print(f"Ignored invalid connection: {src} <-> {dst}")
                except ValueError:
                    print(f"Ignored invalid line: {line.strip()}")

    def _str_(self):
        """Returns a string representation of the adjacency matrix."""
        result = ""
        for i in range(self._rooms):
            result += f"{i}: {' '.join(map(str, sorted(self._matrix[i])))}\n"
        return result.strip()

    def is_connected(self):
        """Check if the graph is fully connected using DFS."""
        visited = [False] * self._rooms

        def dfs(v):
            visited[v] = True
            for neighbor in self._matrix[v]:
                if not visited[neighbor]:
                    dfs(neighbor)

        dfs(0)
        return all(visited)

    def find_path(self, start, end):
        """Finds a path between two rooms using DFS."""
        path = []
        visited = [False] * self._rooms

        def dfs(v):
            if v == end:
                path.append(v)
                return True
            visited[v] = True
            path.append(v)
            for neighbor in self._matrix[v]:
                if not visited[neighbor]:
                    if dfs(neighbor):
                        return True
            path.pop()
            return False

        if dfs(start):
            return path
        return None

def open_file():
    while True:
        filename = input("Enter a file name: ")
        try:
            open(filename.strip(), "r").close()  # Just checking if it opens
            return filename.strip()
        except FileNotFoundError:
            print(f"File {filename} not found. Please try again.")

def main():
    filename = open_file()
    with open(filename, "r") as file:
        matrix = Matrix()
        matrix.read_file(file)

    print("\nGraph Representation (Adjacency List):")
    print(matrix)

    if matrix.is_connected():
        print("The map is fully connected.")
    else:
        print("The map is NOT fully connected.")

    while True:
        try:
            start = int(input("\nEnter a starting room number: "))
            end = int(input("Enter a destination room number: "))
            if 0 <= start < matrix._rooms and 0 <= end < matrix._rooms:
                path = matrix.find_path(start, end)
                if path:
                    print(f"Path found: {' -> '.join(map(str, path))}")
                else:
                    print("No path found.")
            else:
                print("Invalid room numbers.")
        except ValueError:
            print("Please enter valid integers.")
        again = input("Do you want to find another path? (y/n): ").lower()
        if again != 'y':
            break

if _name_ == "_main_":
    main()
```
